# Abstract and Intro 

> In this work we address the problem of finding reliable pixel-level correspondences under difficult imaging conditions.

- Pixel-level correspondences is the fundamental task to solve many relevant problems in geometric computer vision like Visual Odometry, Structure from Motion, Localization, Mapping, Optical Flow, ... 


## Traditional Approaches 

- Rely on manually engineered features detection and description algorithms 
- Consist of the following pipeline 

The 3 steps pipeline 

1. Feature Detection 
2. Feature Description 
3. Features Matching 


## Proposed Approach 

- Rely on automatically learned features (no manual engineering)
- Consists of a 2 steps pipeline 

1. Joint Features Detection and Description 
2. Features Matching 





## Features Description  

The Features Description is a mapping like 

$$ f(u,v,I_{N}) \rightarrow s \in S $$

with 
- $u,v$ : Feature Image Coordinates 
- $I_{N}$ : Feature Center Neighborhood, representing the appearance (e.g. a $W \times H$ neighborhood)
- $S$ : Mixed Spatial and Semantic Space resulting from cartesian product of $W \times H$ Spatial Domain and a $C$ Semantic Domain depending on the feature description 

**NOTE**: 
- The $S$ is essentially a $W \times H \times C$ Tensor Space, so the bread and butter of Deep Neural Networks hence of CNN as well 
- However at the same time it also represents the output space of manually engineered feature detection and description pipeline
